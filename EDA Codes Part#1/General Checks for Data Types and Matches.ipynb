```python
import joblib
import re
from collections import defaultdict

# Load the pickle file
file_path = '/Users/n7/Desktop/ie University SAMBD Acadamics/Capstone Project/Machine Learning Codes/Trained Models/news_encoder.pkl'
news_map = joblib.load(file_path)

# Function to standardize news IDs
def standardize_news_id(nid):
    """
    Standardize news ID by:
    1. Removing any surrounding brackets or quotes.
    2. Stripping whitespace.
    3. Removing prefix "N" if present.
    """
    nid = re.sub(r"[\[\]'\" ]", "", str(nid))  # Remove unwanted characters
    return nid.lstrip("N")  # Remove leading 'N' if present

# --------- Standardize news_map keys, ensuring unique mappings ---------
standardized_news_map = {}
duplicate_keys = defaultdict(list)

for original_key, v in news_map.items():
    standardized_key = standardize_news_id(original_key)
    if standardized_key not in standardized_news_map:
        standardized_news_map[standardized_key] = v
    else:
        # Track duplicates for possible manual review
        duplicate_keys[standardized_key].append(original_key)

# Report duplicates for review if needed
if duplicate_keys:
    print("Duplicate standardized keys found. Only the first occurrence is used; other entries are ignored.")
    for key, original_keys in duplicate_keys.items():
        print(f"Duplicated Key (Standardized): {key}, Original Keys: {original_keys}")

# Debugging information
print(f"Original number of news IDs in news_map: {len(news_map)}")
print(f"Unique, standardized news IDs in news_map: {len(standardized_news_map)}")

# Overwrite `news_map` with the standardized and unique version
news_map = standardized_news_map

print("Total unique standardized news IDs:", len(news_map))
```

```python
import pandas as pd
import re
import joblib
from collections import defaultdict

# Paths to the CSV and pickle files
csv_path = '/Users/n7/Desktop/ie University SAMBD Acadamics/Capstone Project/Data/MINDlarge_dev/Cleaned Dataset/cleaned_behavior_dataset.csv'
pkl_path = '/Users/n7/Desktop/ie University SAMBD Acadamics/Capstone Project/Machine Learning Codes/Trained Models/news_encoder.pkl'

# Function to standardize news IDs
def standardize_news_id(nid):
    """
    Standardize news ID by:
    1. Removing any surrounding brackets or quotes.
    2. Stripping whitespace.
    3. Removing prefix "N" if present.
    """
    nid = re.sub(r"[\[\]'\" ]", "", str(nid))  # Remove unwanted characters
    return nid.lstrip("N")  # Remove leading 'N' if present

# Load the CSV file and process 'Displayed News List' column
df = pd.read_csv(csv_path)
news_ids_column1 = df['Displayed News List'].str.split(',').explode().apply(standardize_news_id).tolist()
all_news_ids_set = set(news_ids_column1)
unique_news_ids_count = len(all_news_ids_set)

# Load the pickle file and process the keys
news_map = joblib.load(pkl_path)
standardized_news_map = {}
duplicate_keys = defaultdict(list)

for original_key, v in news_map.items():
    standardized_key = standardize_news_id(original_key)
    if standardized_key not in standardized_news_map:
        standardized_news_map[standardized_key] = v
    else:
        duplicate_keys[standardized_key].append(original_key)

# Update the news_map with standardized, unique keys
news_map_standardized_keys = set(standardized_news_map.keys())

# Calculate the intersection of standardized IDs from both sources
intersection_ids = all_news_ids_set.intersection(news_map_standardized_keys)

# Outputs
print(f"Total unique news IDs in CSV: {unique_news_ids_count}")
print(f"Total unique standardized news IDs in news_map: {len(news_map_standardized_keys)}")
print(f"Intersection count: {len(intersection_ids)}")
print("Sample intersection IDs:", list(intersection_ids)[:10])
```
